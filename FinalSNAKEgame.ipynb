```python

import pygame
import random
import sys

#initialize pygame
pygame.init()

#define variables, and screen 
WIDTH, HEIGHT = 1000, 600
BLOCK_SIZE = 20
WINDOW_NAME = 'snake sim 2026'
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption(WINDOW_NAME)
pygame.display.set_icon(icon)
clock = pygame.time.Clock()

#define colors
WHITE = (255, 255, 255)
GREEN = (21, 125, 52)
RED = (200, 0, 0)
BLUE = (153,200,250)
GREY = (128,128,128)
BLACK = (0,0,0)

#define font and count
FONT = pygame.font.SysFont(None, 36)

count = 3

#draws the snake, food and block
def draw_snake(snake):
    for segment in snake:
        pygame.draw.rect(screen, BLUE, (*segment, BLOCK_SIZE, BLOCK_SIZE))

def draw_food(food):
    pygame.draw.rect(screen, RED, (*food, BLOCK_SIZE, BLOCK_SIZE))

def draw_blocks(blocks):
    for block in blocks:
        pygame.draw.rect(screen, GREY, (*block, BLOCK_SIZE, BLOCK_SIZE))

#spawns food and ensures that it does not spawn on snake
def food_spawn(snake):
    while True:
        food_pos = (random.randrange(0, WIDTH, BLOCK_SIZE),
                    random.randrange(0, HEIGHT, BLOCK_SIZE))
        if food_pos not in snake:
            return food_pos

#spawn block and ensure it doesnt spawn on food or snake
def block_spawn(snake, food, count):
    blocks = []
    while len(blocks) < count:
        block_pos = (random.randrange(0, WIDTH, BLOCK_SIZE),
                     random.randrange(0, HEIGHT, BLOCK_SIZE))
        if block_pos not in snake and block_pos != food and block_pos not in blocks:
            blocks.append(block_pos)
    return blocks

def ob_one(snake):
    choice1 = ''
    while True:
        screen.fill(GREEN)
        line_1 = FONT.render("Oh wow you are so big already. But I bet you can get bigger.",True,WHITE)  
        line_2 = FONT.render("You just have to consume me with your snake mouth.",True,WHITE)
        line_3 = FONT.render("What will you do: 1. Eat the berry, 2. Dont eat the berry.", True, WHITE)
        screen.blit(line_1, (WIDTH//2-380, HEIGHT//2 - 40)) 
        screen.blit(line_2, (WIDTH//2-380, HEIGHT//2 - 10))
        screen.blit(line_3, (WIDTH//2-380, HEIGHT//2 + 20))
        pygame.display.flip()
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit(); sys.exit()
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_1:
                    choice1 = 'DONT_EAT'
                    while True:
                        screen.fill(GREEN)
                        option_2 = FONT.render("You ate the berry but it was poisoned, no length gained.", True, WHITE)
                        RETURN_1 = FONT.render("Press R to return to game", True, WHITE)
                        screen.blit(option_2, (WIDTH//2-320, HEIGHT//2 - 40))
                        screen.blit(RETURN_1, (WIDTH//2-150, HEIGHT//2 - 10))
                        pygame.display.flip()

                        for event in pygame.event.get():
                            if event.type == pygame.QUIT:
                                pygame.quit(); sys.exit()
                            elif event.type == pygame.KEYDOWN:
                                if event.key == pygame.K_r:
                                    return choice1
                        
                elif event.key == pygame.K_2:
                    choice1 = 'DONT_EAT'
                    while True:
                        screen.fill(GREEN)
                        option_3 = FONT.render("'Oh no you don’t want to eat me. It’s okay, I wouldn’t want to eat me either, ", True, WHITE)
                        option_3_2 = FONT.render("given I'm poisoned and all.'", True, WHITE)
                        RETURN_2 = FONT.render("Press R to return to game", True, WHITE)
                        screen.blit(option_3, (WIDTH//2-420, HEIGHT//2 - 40))
                        screen.blit(option_3_2, (WIDTH//2-170, HEIGHT//2 - 10))
                        screen.blit(RETURN_2, (WIDTH//2-160, HEIGHT//2 + 20))
                        pygame.display.flip()

                        for event in pygame.event.get():
                            if event.type == pygame.QUIT:
                                pygame.quit(); sys.exit()
                            elif event.type == pygame.KEYDOWN:
                                if event.key == pygame.K_r:
                                   return choice1
def ob_two(snake):                                
    choice2 = ''
    while True:
        screen.fill(GREEN)
        line_1 = FONT.render("You found the magical super berry!!!",True,WHITE)  
        line_2 = FONT.render("What will you do: 1. Eat the berry, 2. Dont eat the berry.",True,WHITE)
        screen.blit(line_1, (WIDTH//2-270, HEIGHT//2 - 40)) 
        screen.blit(line_2, (WIDTH//2-350, HEIGHT//2 - 10))
        pygame.display.flip()
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit(); sys.exit()
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_1:
                    choice2 = 'DOUBLE'
                    
                    while True:
                        screen.fill(GREEN)
                        option_2 = FONT.render("You ate the berry, length doubled.", True, WHITE)
                        RETURN_1 = FONT.render("Press R to return to game", True, WHITE)
                        screen.blit(option_2, (WIDTH//2-200, HEIGHT//2 - 40))
                        screen.blit(RETURN_1, (WIDTH//2-150, HEIGHT//2 - 10))
                        pygame.display.flip()

                        for event in pygame.event.get():
                            if event.type == pygame.QUIT:
                                pygame.quit(); sys.exit()
                            elif event.type == pygame.KEYDOWN:
                                if event.key == pygame.K_r:
                                    return choice2
                 
                elif event.key == pygame.K_2:
                    choice2 = 'DONT_EAT'
                    while True:
                        screen.fill(GREEN)
                        option_3 = FONT.render("'What a Shame'", True, WHITE)
                        option_3_2 = FONT.render("'You're missing out on a great meal'", True, WHITE)
                        RETURN_2 = FONT.render("Press R to return to game", True, WHITE)
                        screen.blit(option_3, (WIDTH//2-130, HEIGHT//2 - 40))
                        screen.blit(option_3_2, (WIDTH//2-190, HEIGHT//2 - 10))
                        screen.blit(RETURN_2, (WIDTH//2-160, HEIGHT//2 + 20))
                        pygame.display.flip()

                        for event in pygame.event.get():
                            if event.type == pygame.QUIT:
                                pygame.quit(); sys.exit()
                            elif event.type == pygame.KEYDOWN:
                                if event.key == pygame.K_r:
                                   return choice2
def ob_three(snake):
    choice3 = ''
    while True:
        screen.fill(GREEN)
        line_1 = FONT.render("You found a normal berry",True,WHITE)  
        line_2 = FONT.render("What will you do: 1. Eat the berry, 2. Dont eat the berry.", True, WHITE)
        screen.blit(line_1, (WIDTH//2-150, HEIGHT//2 - 40)) 
        screen.blit(line_2, (WIDTH//2-300, HEIGHT//2 - 10))
        pygame.display.flip()
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit(); sys.exit()
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_1:
                    choice3 = 'EAT'
                    while True:
                        screen.fill(GREEN)
                        option_2 = FONT.render("You ate the berry, length added.", True, WHITE)
                        RETURN_1 = FONT.render("Press R to return to game", True, WHITE)
                        screen.blit(option_2, (WIDTH//2-170, HEIGHT//2 - 40))
                        screen.blit(RETURN_1, (WIDTH//2-150, HEIGHT//2 - 10))
                        pygame.display.flip()

                        for event in pygame.event.get():
                            if event.type == pygame.QUIT:
                                pygame.quit(); sys.exit()
                            elif event.type == pygame.KEYDOWN:
                                if event.key == pygame.K_r:
                                    return choice3
                        
                elif event.key == pygame.K_2:
                    choice3 = 'DONT_EAT'
                    while True:
                        screen.fill(GREEN)
                        option_3 = FONT.render("'Oh no you don’t want to eat me. So sad :('", True, WHITE)
                        RETURN_2 = FONT.render("Press R to return to game", True, WHITE)
                        screen.blit(option_3, (WIDTH//2-220, HEIGHT//2 - 40))
                        screen.blit(RETURN_2, (WIDTH//2-160, HEIGHT//2 - 10))
                        pygame.display.flip()

                        for event in pygame.event.get():
                            if event.type == pygame.QUIT:
                                pygame.quit(); sys.exit()
                            elif event.type == pygame.KEYDOWN:
                                if event.key == pygame.K_r:
                                   return choice3

def ob_four(snake):
    choice4 = ''
    while True:
        screen.fill(GREEN)
        line_1 = FONT.render("You found the Special Chris berry",True,WHITE)  
        line_2 = FONT.render("Chris is hungry and will eat some of you",True,WHITE)
        line_3 = FONT.render("What will you do: 1. Eat the berry, 2. Dont eat the berry.", True, WHITE)
        screen.blit(line_1, (WIDTH//2-200, HEIGHT//2 - 40)) 
        screen.blit(line_2, (WIDTH//2-240, HEIGHT//2 - 10))
        screen.blit(line_3, (WIDTH//2-310, HEIGHT//2 + 20))
        pygame.display.flip()
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit(); sys.exit()
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_1:
                    choice4 = 'DONT_EAT'
                    while True:
                        screen.fill(GREEN)
                        option_2 = FONT.render("You ate the special Chris berry, but he ate some of you, no length gained.", True, WHITE)
                        RETURN_1 = FONT.render("Press R to return to game", True, WHITE)
                        screen.blit(option_2, (WIDTH//2-420, HEIGHT//2 - 40))
                        screen.blit(RETURN_1, (WIDTH//2-150, HEIGHT//2 - 10))
                        pygame.display.flip()

                        for event in pygame.event.get():
                            if event.type == pygame.QUIT:
                                pygame.quit(); sys.exit()
                            elif event.type == pygame.KEYDOWN:
                                if event.key == pygame.K_r:
                                    return choice4
                        
                elif event.key == pygame.K_2:
                    choice4 = 'DONT_EAT'
                    while True: 
                        screen.fill(GREEN)
                        option_3 = FONT.render("Chris is not happy that you chose not to eat him ", True, WHITE)
                        RETURN_2 = FONT.render("Press R to return to game", True, WHITE)
                        screen.blit(option_3, (WIDTH//2-300, HEIGHT//2 - 40))
                        screen.blit(RETURN_2, (WIDTH//2-160, HEIGHT//2 - 10))
                        pygame.display.flip()

                        for event in pygame.event.get():
                            if event.type == pygame.QUIT:
                                pygame.quit(); sys.exit()
                            elif event.type == pygame.KEYDOWN:
                                if event.key == pygame.K_r:
                                   return choice4

def ob_five(snake):
    choice5 = ''
    while True:
        screen.fill(GREEN)
        line_1 = FONT.render("'It would be a bummer if some one eat me up!",True,WHITE)  
        line_2 = FONT.render("Being eaten is my least favorite thing ever!'",True,WHITE)
        line_3 = FONT.render("What will you do: 1. eat him, 2. dont eat him", True, WHITE)
        screen.blit(line_1, (WIDTH//2-380, HEIGHT//2 - 40)) 
        screen.blit(line_2, (WIDTH//2-380, HEIGHT//2 - 10))
        screen.blit(line_3, (WIDTH//2-380, HEIGHT//2 + 20))
        pygame.display.flip()
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit(); sys.exit()
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_1:
                    choice1 = 'DOUBLE'
                    while True:
                        screen.fill(GREEN)
                        option_6 = FONT.render("You ate the berry +2 length", True, WHITE)
                        RETURN_5 = FONT.render("Press R to return to game", True, WHITE)
                        screen.blit(option_6, (WIDTH//2-150, HEIGHT//2 - 40))
                        screen.blit(RETURN_5, (WIDTH//2-150, HEIGHT//2 - 10))
                        pygame.display.flip()

                        for event in pygame.event.get():
                            if event.type == pygame.QUIT:
                                pygame.quit(); sys.exit()
                            elif event.type == pygame.KEYDOWN:
                                if event.key == pygame.K_r:
                                    return choice5
                        
                elif event.key == pygame.K_2:
                    choice1 = 'DONT_EAT'
                    while True:
                        screen.fill(GREEN)
                        option_7 = FONT.render("'Im so glad no one ate me!", True, WHITE)
                        option_7_2 = FONT.render(" I wouldnt like that at all!.'", True, WHITE)
                        RETURN_6 = FONT.render("Press R to return to game", True, WHITE)
                        screen.blit(option_7, (WIDTH//2-170, HEIGHT//2 - 40))
                        screen.blit(option_7_2, (WIDTH//2-170, HEIGHT//2 - 10))
                        screen.blit(RETURN_2, (WIDTH//2-160, HEIGHT//2 + 20))
                        pygame.display.flip()

                        for event in pygame.event.get():
                            if event.type == pygame.QUIT:
                                pygame.quit(); sys.exit()
                            elif event.type == pygame.KEYDOWN:
                                if event.key == pygame.K_r:
                                    return choice5

def ob_six(snake):
    choice1 = ''
    while True:
        screen.fill(GREEN)
        line_1 = FONT.render("'I wish there was a snake here that would eat me up!",True,WHITE)  
        line_2 = FONT.render("Its a shame snakes havent been seen in these parts in years!!'",True,WHITE)
        line_3 = FONT.render("What will you do: 1. eat him, 2. Dont keat him", True, WHITE)
        screen.blit(line_1, (WIDTH//2-380, HEIGHT//2 - 40)) 
        screen.blit(line_2, (WIDTH//2-380, HEIGHT//2 - 10))
        screen.blit(line_3, (WIDTH//2-380, HEIGHT//2 + 20))
        pygame.display.flip()
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit(); sys.exit()
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_1:
                    choice1 = 'DOUBLE'
                    while True:
                        screen.fill(GREEN)
                        option_4 = FONT.render("'Oh wow! a snaking! eating me! Yolly good show'", True, WHITE)
                        option_4_2 = FONT.render("You ate the berry. +2", True, WHITE)
                        RETURN_3 = FONT.render("Press R to return to game", True, WHITE)
                        screen.blit(option_4, (WIDTH//2-220, HEIGHT//2 - 40))
                        screen.blit(option_4_2, (WIDTH//2-170, HEIGHT//2 - 10))
                        screen.blit(RETURN_3, (WIDTH//2-160, HEIGHT//2 + 20))
                        pygame.display.flip()

                        for event in pygame.event.get():
                            if event.type == pygame.QUIT:
                                pygame.quit(); sys.exit()
                            elif event.type == pygame.KEYDOWN:
                                if event.key == pygame.K_r:
                                    return choice1
                        
                elif event.key == pygame.K_2:
                    choice1 = 'DONT_EAT'
                    while True:
                        screen.fill(GREEN)
                        option_5 = FONT.render("'Huh. I really thought that would make the snake watching me eat me.", True, WHITE)
                        option_5_2 = FONT.render(" Welp what can ya do.'", True, WHITE)
                        RETURN_4 = FONT.render("Press R to return to game", True, WHITE)
                        screen.blit(option_5, (WIDTH//2-400, HEIGHT//2 - 40))
                        screen.blit(option_5_2, (WIDTH//2-170, HEIGHT//2 - 10))
                        screen.blit(RETURN_4, (WIDTH//2-160, HEIGHT//2 + 20))
                        pygame.display.flip()

                        for event in pygame.event.get():
                            if event.type == pygame.QUIT:
                                pygame.quit(); sys.exit()
                            elif event.type == pygame.KEYDOWN:
                                if event.key == pygame.K_r:
                                    return choice1

def ob_seven(snake):
    choice1 = ''
    while True:
        screen.fill(GREEN)
        line_1 = FONT.render("'I wish there was a snake here that would eat me up!",True,WHITE)  
        line_2 = FONT.render("Its a shame snakes havent been seen in these parts in years!!'",True,WHITE)
        line_3 = FONT.render("What will you do: 1. eat him, 2. Dont keat him", True, WHITE)
        screen.blit(line_1, (WIDTH//2-380, HEIGHT//2 - 40)) 
        screen.blit(line_2, (WIDTH//2-380, HEIGHT//2 - 10))
        screen.blit(line_3, (WIDTH//2-380, HEIGHT//2 + 20))
        pygame.display.flip()
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit(); sys.exit()
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_1:
                    choice1 = 'DOUBLE'
                    while True:
                        screen.fill(GREEN)
                        option_4 = FONT.render("'Oh wow! a snaking! eating me! Yolly good show'", True, WHITE)
                        option_4_2 = FONT.render("You ate the berry. +2", True, WHITE)
                        RETURN_3 = FONT.render("Press R to return to game", True, WHITE)
                        screen.blit(option_4, (WIDTH//2-200, HEIGHT//2 - 40))
                        screen.blit(option_4_2, (WIDTH//2-170, HEIGHT//2 - 10))
                        screen.blit(RETURN_3, (WIDTH//2-160, HEIGHT//2 + 20))
                        pygame.display.flip()

                        for event in pygame.event.get():
                            if event.type == pygame.QUIT:
                                pygame.quit(); sys.exit()
                            elif event.type == pygame.KEYDOWN:
                                if event.key == pygame.K_r:
                                    return choice1
                        
                elif event.key == pygame.K_2:
                    choice1 = 'DONT_EAT'
                    while True:
                        screen.fill(GREEN)
                        option_5 = FONT.render("'Huh. I really thought that would make the snake watching me eat me.", True, WHITE)
                        option_5_2 = FONT.render(" Welp what can ya do.'", True, WHITE)
                        RETURN_4 = FONT.render("Press R to return to game", True, WHITE)
                        screen.blit(option_5, (WIDTH//2-350, HEIGHT//2 - 40))
                        screen.blit(option_5_2, (WIDTH//2-170, HEIGHT//2 - 10))
                        screen.blit(RETURN_4, (WIDTH//2-160, HEIGHT//2 + 20))
                        pygame.display.flip()

                        for event in pygame.event.get():
                            if event.type == pygame.QUIT:
                                pygame.quit(); sys.exit()
                            elif event.type == pygame.KEYDOWN: 
                                if event.key == pygame.K_r:
                                    return choice1

def ob_eight(snake):
    choice1 = ''
    while True:
        screen.fill(GREEN)
        line_1 = FONT.render("'I bet i could beat any snake that tried to eat me!",True,WHITE)  
        line_2 = FONT.render("It wouldnt stand a chance'",True,WHITE)
        line_3 = FONT.render("What will you do: 1. eat him, 2. Dont eat him", True, WHITE)
        screen.blit(line_1, (WIDTH//2-380, HEIGHT//2 - 40)) 
        screen.blit(line_2, (WIDTH//2-340, HEIGHT//2 - 10))
        screen.blit(line_3, (WIDTH//2-380, HEIGHT//2 + 20))
        pygame.display.flip()
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit(); sys.exit()
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_1:
                    choice1 = 'DOUBLE'
                    while True:
                        screen.fill(GREEN)
                        option_8 = FONT.render("'Oh no i was so wrong, a snake would beat me!'", True, WHITE)
                        option_8_2 = FONT.render("You ate the berry. +2", True, WHITE)
                        RETURN_7 = FONT.render("Press R to return to game", True, WHITE)
                        screen.blit(option_8, (WIDTH//2-250, HEIGHT//2 - 40))
                        screen.blit(option_8_2, (WIDTH//2-170, HEIGHT//2 - 10))
                        screen.blit(RETURN_7, (WIDTH//2-160, HEIGHT//2 + 20))
                        pygame.display.flip()

                        for event in pygame.event.get():
                            if event.type == pygame.QUIT:
                                pygame.quit(); sys.exit()
                            elif event.type == pygame.KEYDOWN:
                                if event.key == pygame.K_r:
                                    return choice1
                        
                elif event.key == pygame.K_2:
                    choice1 = 'DONT_EAT'
                    while True:
                        screen.fill(GREEN)
                        option_9 = FONT.render("Yeah no snack would make me a snack", True, WHITE)
                        option_9_2 = FONT.render(" They're just a gov psy-op to keep us in'", True, WHITE)
                        RETURN_8 = FONT.render("Press R to return to game", True, WHITE)
                        screen.blit(option_9, (WIDTH//2-180, HEIGHT//2 - 40))
                        screen.blit(option_9_2, (WIDTH//2-170, HEIGHT//2 - 10))
                        screen.blit(RETURN_8, (WIDTH//2-160, HEIGHT//2 + 20))
                        pygame.display.flip()

                        for event in pygame.event.get():
                            if event.type == pygame.QUIT:
                                pygame.quit(); sys.exit()
                            elif event.type == pygame.KEYDOWN:
                                if event.key == pygame.K_r:
                                    return choice1

def random_ob(snake):
    choice = random.choice([ob_one, ob_two, ob_three, ob_four, ob_five, ob_six, ob_seven, ob_eight])
    return choice(snake)

#game over screen, called when snake dies (hits wall or block or self) and prompts user to restart or quit game
def game_over(score):
    while True:
        screen.fill(BLACK)
        lose_text = FONT.render(f"Final Score: {score}", True, WHITE)
        restart_text = FONT.render("Press M to return to main menu, or Q to quit.", True, WHITE)
        screen.blit(lose_text, (WIDTH//2-95, HEIGHT//2 - 40))
        screen.blit(restart_text, (WIDTH//2-245, HEIGHT//2 + 10))
        pygame.display.flip()
    
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit(); sys.exit()
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_m:
                    return True
                elif event.key == pygame.K_q:
                    pygame.quit(); sys.exit()

def main_menu():
    while True:
        screen.fill(BLACK)
        menu_text = FONT.render("Welcome to Snake 2!", True, WHITE)
        play_text = FONT.render("Press P to play", True, WHITE)
        exit_game = FONT.render("Press Q to exit to desktop", True, WHITE)
        screen.blit(menu_text, (WIDTH//2-130, HEIGHT//2 - 120))
        screen.blit(play_text, (WIDTH//2-100, HEIGHT//2 - 50))
        screen.blit(exit_game, (WIDTH//2-150, HEIGHT//2 +10))
        pygame.display.flip()

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit(); sys.exit()
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_p:
                    game()
                elif event.key == pygame.K_q:
                    pygame.quit(); sys.exit()

#main game function so that we can have different instances of the game  
def game():
    #initilizing variables
    score = 0
    snake = [(100, 100), (80, 100), (60, 100)] #snake is a list of tuples, each tuple is segemnt of the body, and they are xy coordinates
    direction = (BLOCK_SIZE, 0) 
    food = food_spawn(snake)
    block_list = block_spawn(snake, food, count)

#detect user input while game is running
    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit(); sys.exit()
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_UP and direction != (0, BLOCK_SIZE):
                    direction = (0, -BLOCK_SIZE)
                elif event.key == pygame.K_DOWN and direction != (0, -BLOCK_SIZE):
                    direction = (0, BLOCK_SIZE)
                elif event.key == pygame.K_LEFT and direction != (BLOCK_SIZE, 0):
                    direction = (-BLOCK_SIZE, 0)
                elif event.key == pygame.K_RIGHT and direction != (-BLOCK_SIZE, 0):
                    direction = (BLOCK_SIZE, 0)
        
        #create and insert new head at first index of the snake
        new_head = (snake[0][0] + direction[0], snake[0][1] + direction[1])
        snake.insert(0, new_head) 


        #if head touches food, randomly pick dialogue, deteremine whether to add to score or pop and respawn vlock and food
        if new_head == food:
            food = food_spawn(snake)
            block_list = block_spawn(snake, food, count)
            choice = random_ob(snake)
            if choice == 'EAT':
                score += 1
            elif choice == 'DOUBLE':
                score += 2
            elif choice == 'DONT_EAT':
                snake.pop()
           
                
        else:
            snake.pop() #ensures the snake doesnt keep growing while moving without eating an apple

        #restart game if snake hits block
        if new_head in block_list:
            if game_over(score):
                return

        #checks if the head touches the borders or any other part of the snake, restarting the game if it does
        if (new_head[0] < 0 or new_head[0] >= WIDTH or
            new_head[1] < 0 or new_head[1] >= HEIGHT or
            new_head in snake[1:]):
            if game_over(score):
                return 

        #fill screen with green and call functions
        screen.fill(GREEN)
        draw_snake(snake)
        draw_food(food)
        draw_blocks(block_list)

        #initilize the score text and place it top left of the screen 
        score_text = FONT.render(f"SCORE: {score}", True, WHITE)
        screen.blit(score_text, (10, 10))

        #keep redrawing the display
        pygame.display.flip()

        #increase movement speed as snake grows longer
        if len(snake) < 8: clock.tick(10)  
        elif len (snake) > 8 and len(snake) < 12: clock.tick(12)
        else: clock.tick(14)

#call main menu
main_menu()
```
